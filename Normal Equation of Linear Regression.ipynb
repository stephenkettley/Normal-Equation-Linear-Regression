## Import modules, create dataset and plot dataset

### Import required modules

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

### Create dataset

```python
x,y = make_regression(n_samples=100, n_features=1,n_informative=1,noise=10,random_state=10)
```

### Plot the generated dataset

```python
plt.scatter(x,y, s=30, marker='o')
plt.xlabel("Feature 1",fontsize=18)
plt.ylabel('Target Variable',fontsize=18)
plt.title('Simple Linear Regression', fontsize=25)
plt.show()
```

### Convert the target variable array from 1d to 2d

```python
y = y.reshape(100,1)
```

## Let's implement the Normal equation

### Adding x0 to each instance

```python
x_new = np.c_[np.ones((100,1)), x]
```

### Using the normal equation

```python
theta_best_values = np.linalg.inv(x_new.T.dot(x_new)).dot(x_new.T).dot(y)
```

### Display the best values obtained

```python
print(theta_best_values)
```

## Try to predict for new data instance

### Sample data instance

```python
x_sample = np.array([[-2],[4]])
```

### Adding x0 to each sample instance

```python
x_sample_new = np.c_[np.ones((2,)), x_sample]
```

### Display the sample

```python
print('Before adding x0:\n', x_sample)
print('After adding x0:\n', x_sample_new)
```

### Predict the values for the given data instance

```python
predict_value = x_sample_new.dot(theta_best_values)
print(predict_value)
```

## Plot the output

```python
plt.scatter(x,y, s=30, marker='o')
plt.plot(x_sample, predict_value, c='red')
plt.plot()
plt.xlabel('Feature 1', fontsize=12)
plt.ylabel('Target Variable', fontsize=12)
plt.title('Simple Linear Regression (with line of best fit)', fontsize=20)
plt.show()
```

## We can verify the above using sklearn LinearRegression class

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

### Print obtained theta values

```python
print('Best value of theta is: ', model.intercept_, model.coef_, sep='\n')
```

### Predict

```python
print('Predicted value: ', model.predict(x_sample), sep='\n')
```
